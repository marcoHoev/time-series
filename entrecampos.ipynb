```python
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import STL
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('data/QualidadeARO3.xlsx')
data = data.drop(columns='Ihavo')
rng = pd.date_range('1/1/2020 00:00', periods=8784, freq='1H')
data.index = rng
```

## 1. Entrecampos Analysis

```python
#entrecampos raw
entrecampos = data.Entrecampos

plt.figure(figsize=(20,20))

plot_pacf(entrecampos, ax=plt.subplot(3, 1, 1), lags=50, title='pacf entrcampos raw')
plot_acf(entrecampos, ax=plt.subplot(3, 1, 2), lags=50, title='acf entrecampos raw')
res = STL(entrecampos).fit()

plt.subplot(3, 1, 3)
res.plot()
plt.show()
```

```python
#entrecampos differencing 1
entrecampos = entrecampos.diff()
entrecampos = entrecampos[1:]
plt.plot(data.index, data['Entrecampos'], color='red')
plt.plot(entrecampos)

#pacf, acf plots
plot_pacf(entrecampos, lags=50, title='pacf entrcampos diff with 1')
plot_acf(entrecampos, lags=50, title='acf entrecampos diff with 1')
#decomposition plot
res = STL(entrecampos).fit()
res.plot()
plt.show()

```

```python
#entrecampos differencing 24
entrecampos = entrecampos.diff(24)
entrecampos = entrecampos[24:]
plt.plot(data.index, data['Entrecampos'], color='red')
plt.plot(entrecampos)

#pacf, acf plots
plot_pacf(entrecampos, lags=50, title='pacf entrcampos diff with 24')
plot_acf(entrecampos, lags=50, title='acf entrecampos diff with 24')
#decomposition plot
res = STL(entrecampos).fit()
res.plot()
plt.show()
```
